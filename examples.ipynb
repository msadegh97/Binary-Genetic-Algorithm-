## Binary Genetic Algorithm 

```python
from BGA import BGA
import numpy as np 
```

```python
def values(arr):
    w= [1, 2, 4 ,8 ,-16] 
    out = np.sum(w * arr)
    return out

```

```python
model = BGA(max_iteratoin=30, n_population=6, num_param=5, score_func=values, prob_crossover=.5, prob_mut=.5, crossover_method='uniform')
```

```python
model.fit()
```

```python
model.print_pop()
```
